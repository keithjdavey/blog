## First of all is a Tic Tac Toe game.

Here, each player is asked in sequence what coordinates they want to place their mark, with 1,1 and 3,3 corresponding to the top left and bottom right corners respectively. After each player makes their choice, the board is shown. If a player gets 3 in a row, the game states the winner and exits the program. If the player makes an invalid move (e.g., in the same spot as another mark), the program asks them to go again.

```python
def checkEqual(lst):
   return lst[1:] == lst[:-1]

def winCheck(game):
    for i in range(len(game)):
        if checkEqual(game[i]) is True and game[i][0] != ' ':
            return(f"Player {game[i][0]} is the winner.")
        elif checkEqual([item[i] for item in game]) is True and game[0][i] != ' ':
            return(f"Player {[item[i] for item in game][0]} is the winner.")
        elif game[1][1] != ' ':
            if game[0][0] == game[1][1] == game[2][2] or game[0][2] == game[1][1] == game[2][0]:
                return(f"Player {game[1][1]} is the winner.")

def printGame(inp):
   vlines = 3*'|   ' + '|'
   X = [vlines.split(),vlines.split(),vlines.split()]
   hlines = ' '+3*'--- '
   print(hlines)
   for i in range(3):
      for j in range(3):
         X[i].insert((2*j)+1, game[i][j])
      print(' '.join(X[i]))
      print(hlines)

game = [[' ', ' ', ' '],[' ', ' ', ' '],[' ', ' ', ' ']]
markers = ['X', 'O']
Running = True

while Running is True:
    for k in range(1,3):
        pcoord = input("Enter coords for placement of marker (row,col): ").split(',')
        if game[int(pcoord[0])-1][int(pcoord[1])-1] != ' ':
            printGame(game)
            pcoord = input("Enter coords for placement of marker (row,col): ").split(',')
        game[int(pcoord[0])-1][int(pcoord[1])-1] = markers[k-1]

        if winCheck(game) != None:
            print(winCheck(game))
            printGame(game)
            Running = False
            break
        elif ' ' not in game[0]+game[1]+game[2]:
            print("Draw!")
            printGame(game)
            Running = False
            break
        else:
            printGame(game)
```

## Below is a game of Hangman

The game capitalises the letter you put in and checks it against the word it chooses randomly from a dictionary text file.

```python
from random import choice

def hangmanInit():
    with open('dictionary.txt', 'r') as open_dict:
        dicti = open_dict.read().split('\n')

    randword = ' '.join(choice(dicti))

    pword = ['_']
    for i in range(int((len(randword)-1)/2)):
        pword.append(' _')
    pword = ''.join(pword)

    guessed = []
    lives = 6

    return(guessed, lives, pword, randword)

def hangmanDraw(lives):
    hang = ['_____', '|   |' , '|   |' , '|    ' , '|      ' , '|       ', '|     ','|      ','|','|_________']

    for i in range(len(hang)):
            hang[i] = list(hang[i])

    if lives <= 5:
        hang[3][-1] = 'O'
        if lives <= 4:
            hang[4][5], hang[4][4] = '|','|'
            hang[5][5], hang[5][4] = '|','|'
            if lives <= 3:
                hang[5][2],hang[4][3] = '/','/'
                if lives <= 2:
                    hang[4][-1],hang[5][-1] = '\\','\\'
                    if lives <= 1:
                        hang[7][3],hang[6][4] = '/','/'
                        if lives == 0:
                            hang[6][5],hang[7][-1] = '\\','\\'

    for i in range(len(hang)):
            hang[i] = ''.join(hang[i])

    for i in range(len(hang)):
        print(hang[i])

X = hangmanInit()
guessed, lives, pword, randword = X[0], X[1], X[2], X[3]

print(f"The word to guess has {int((len(randword)+1)/2)} letters.")

while pword != randword and lives != 0:
    guessed.sort()
    if len(guessed) > 0:
        print(f"The letters you have already guessed are as follows:\n{guessed}")
    plet = (input("Please enter a letter to guess: ")).upper()
    while len(plet) != 1 or plet.isupper() is False or plet.upper() in guessed:
        plet = (input("Please enter a letter to guess: ")).upper()
    guessed.append(plet.upper())

    if plet.upper() in randword:
        indices = [i for i, x in enumerate(randword) if x == plet.upper()]
        pword = list(pword)
        for j in range(len(indices)):
            pword[indices[j]] = plet.upper()
        pword = ''.join(pword)
    else:
        lives -= 1
        print(f'\nIncorrect, you have {lives} choices left.')
    
    print(f"{pword}\n")
    hangmanDraw(lives)
    if pword == randword: print("\nCongratulations, you guessed the word!")
    if lives == 0: print("\n You have run out of lives.")
    if pword == randword or lives == 0:
        print(f"The word was {randword}.\n")
        retry = input("Would you like to try again? Type Y for yes or N for no: ")
        if retry.upper() == 'Y':
            X = hangmanInit()
            guessed, lives, pword, randword = X[0], X[1], X[2], X[3]
            print(f"The word to guess has {int((len(randword)+1)/2)} letters.")
        elif retry.upper() != 'N':
            while retry.upper() != 'N':
                retry = input("Would you like to try again? Type Y for yes or N for no: ")
```
